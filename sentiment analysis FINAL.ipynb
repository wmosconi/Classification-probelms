```python
f = open("/Users/williammosconiii/Desktop/sentiment-labelled-sentences/imdb_labelled.txt", 'r')
data = []
for line in f:
    line = line.replace('\n', '').strip().replace('\t', '')
    line = line.replace('?', '')
    line = line.replace('!', '')
    line = line.replace('.', '')
    line = line.replace('-', '')
    line = line.replace("'", '')
    line = line.replace(',', '')
    line = line.replace('"', '')
    line = line.replace(')', '')
    line = line.replace('(', '')
    line = line.replace(':', '')
    line = line.replace(';', '')
    data.append({'Review':line[:-1], 'Sentiment':line[-1:]})
```

```python
df = pd.DataFrame.from_dict(data)
pd.set_option('display.max_colwidth', -1)
df
```
